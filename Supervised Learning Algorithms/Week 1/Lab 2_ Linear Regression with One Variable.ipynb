## Libraries
- **numpy**: library to deal with arrays.
- **matplotlib**: library for plotting data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data Set
- Our data set will be consists of only two data points.
- Save the number of examples in the varaible **m**.

```python
# The input variabl or Y.
x_train = np.array([1.0, 2.0])

# The target or the output variable Y.
y_train = np.array([300.0, 500.0])

# We can calculate the value of m by two ways:
# Using .shape parameter.
m = x_train.shape[0]

# Using len() method.
m = len(x_train)

```

## Plotting the Data
- The method **scatter()** from the **matplotlib** library used to plot the data as below.
It takes two parameter:
    1. **marker** to specify the character to be drawn in the plot in the place of the data point.
    2. **c** to specify the color of the data point.
- After that we specify the title of the plot, the **x** axis and the same for **y** axis and finally show the plot.

```python
plt.scatter(x_train, y_train, marker='o', c='b')
plt.title('Housing Prices')
plt.ylabel('Price in 1000s of Dollars')
plt.xlabel('Size in 1000 sqft')
plt.show()
```

## Model Function
- First we specify the values of **w** and **b**.
- Then define a function to calculate the value for the data points.
- Finally, we call the function and plot out output.

```python
# Change them and see the difference.
w = 200
b = 100

def compute_model_output(x, w, b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b
  return f_wb

f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, f_wb, c='r', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='b', label='Actual Values')
plt.title('Housing Prices')
plt.ylabel('Price in 1000s of Dollars')
plt.xlabel('Size in 1000 sqft')
plt.legend()
plt.show()
```

## Make your own prediction

```python
print('Enter the size of the house please (in 1000 sqft): ')
user_size = float(input())
prediction = w * user_size + b
print(f'The predicted price is: ${prediction:.0f} thousand dollars')
```
